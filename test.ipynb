```python
import pandas as pd
```

```python
#read the csv file
df_population_raw = pd.read_csv("population_total.csv")
```

```python
#first look at the dataset
df_population_raw
```

```python
#dropping null values
df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot(index="year", columns="country", values="population")
```

```python
#selecting some countries
df_pivot = df_pivot[['Angola', 'India', 'Brazil', 'Yemen', 'Zambia', 'Uzbekistan']]
```

```python
#showing pivot table
df_pivot
```

```python
#Lineplot
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population (1955-2020)', figsize=(18,14))
```

```python
#selecting only one year(2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
#showing new dataframe
df_pivot_2020
```

```python
#transpose dataframe from rows to columns or viceversa
df_pie = df_pivot_2020.T
```

```python
df_pie
```

```python
#making barplot
df_pivot_2020.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population 2020')
```

```python
#selecting a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010,2020])]
```

```python
#making grouped barplot
df_pivot_sample.plot(kind='bar', xlabel='Year', ylabel='Population')
```

```python
#changing column name
df_pie
```

```python
#change the column name
df_pie.rename(columns={2020:'2020'}, inplace=True)
```

```python
df_pie
```
